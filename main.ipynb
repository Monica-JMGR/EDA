```python
import pandas as pd
from plotly.offline import iplot
import plotly.graph_objs as go
import matplotlib.pyplot as plt
```

```python
# Cargamos el archivo excel descargado por la hoja de Ejecuciones Hipotecarias

EH_excel = pd.read_excel('data/Series efecto de la crisis en los organos judiciales por provincias.xlsx', sheet_name= ' Ej.Hipot por provincias')
EH_excel
```

```python
# Usamos la columna 1 como indice que es donde se encuentran las provincias

EH_excel = EH_excel.set_index('Unnamed: 1')
```

```python
# Solo nos quedamos con las filas y columnas que son necesarias para nuestro análisis

EH_excel = EH_excel.iloc[4:54,62:]
EH_excel
```

```python
# Cambiamos el nombre de las columnas

EH_excel.columns = ['2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']
```

```python
# Mostramos el DataFrame

EH_excel
```

```python
# Cambiamos el nombre del indice

EH_excel.index.names = ['Provincias']
```

```python
# Mostramos el DataFrame

EH_excel
```

```python
# Realizamos una gráfica de barras para ver la evolución anual de las Ejecuciones Hipotecarias presentadas

EH_1 = {
    'x': EH_excel.columns,
    'y': EH_excel.sum(),
    'name': 'EH',
    'type': 'bar'
};
data = [EH_1];
layout = {
    'xaxis': {'title': 'Años'},
    'title': 'Ejecuciones Hipotecarias presentadas 2008-2021'
};
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

# Comparativa Ejecuciones Hipotecarias vs. Lanzamientos Judiciales

```python
# Creamos un nuevo DataFrame uniendo las ejecuciones hipotecarias y los lanzamientos judiciales para poder hacer una comparativa anual

EH_LZ = pd.DataFrame({'Ejecuciones Hipotecarias': EH_excel.sum(), 'Lanzamientos Judiciales': LZ_excel.sum()}, index=EH_excel.columns)
EH_LZ
```

```python
# Mostramos los datos del DataFrame en una Gráfica para poder ver la evolución

plt.figure(figsize=(15,5))

x = EH_LZ.index
y1 = EH_LZ['Ejecuciones Hipotecarias']
y2 = EH_LZ['Lanzamientos Judiciales']

plt.plot(x,y1,'-o',label='Ejecuciones Hipotecarias')
plt.plot(x,y2, '-*', label='Lanzamientos')

plt.title("Relación demandas EH presentadas vs. Lanzamientos realizados",fontsize=13)
plt.xlabel("Años",fontsize=10)
plt.ylabel("Nº demandas EH y Lanzamientos",fontsize=10)
plt.ylim(0,100000)
plt.legend()
plt.show();
```

```python
# Trasponemos el DataFrame de las Ejecuciones Hipotecarias para tener en las columnas las provincias

EH = EH_excel.T
EH
```

```python
# Trasponemos el DataFrame de los Lanzamientos para tener en las columnas las provincias

LZ = LZ_excel.T
LZ
```

```python
# Creamos un nuevo DataFrame con los datos por provincias para poder ver la evolución geográfica

EH_LZ_prov = pd.DataFrame({'Ejecuciones Hipotecarias': EH.sum(), 'Lanzamientos Judiciales': LZ.sum()}, index=EH.columns)
EH_LZ_prov
```

```python
# Pintamos en una gráfica los datos para percibir la evolución geográfica

plt.figure(figsize=(18,7))

x = EH_LZ_prov.index
y1 = EH_LZ_prov['Ejecuciones Hipotecarias']
y2 = EH_LZ_prov['Lanzamientos Judiciales']

plt.plot(x,y1,'-o',label='Ejecuciones Hipotecarias')
plt.plot(x,y2, '-*', label='Lanzamientos')

plt.title("Demandas EH presentadas vs. Lanzamientos realizados por provincias",fontsize=13)
plt.xlabel("Provincias",fontsize=10)
plt.ylabel("Nº demandas EH y lanzamientos",fontsize=10)
plt.ylim(0,150000)
plt.xticks(rotation=90)
plt.legend()
plt.show();
```
